1. Write down the numbers from 2 to `n` on a piece of paper.  Call this list of numbers WORKING.

````python
def primes(n):
  p = list(range(2,n+1))
  for i in range(2,n+1):
    for j in range(2,int(i**0.5)+1):
      if i%j == 0:
        p.remove(i)
        break
  return(p)
n = int(input("Enter the value: "))
print(primes(n))
````

and return it as a string.
Use the letters A-Z to accommodate up to base 36.

````python
# Put your code here.
def base_convert(n, b):
    if n == 0:
        return '0'
    converted = ''
    while n:
        converted = str(int(n % b))+converted
        n = int(n/b)
    return converted
n = int(input("Enter the integer "))
b = int(input("Base to be converted to? "))
print(base_convert(n,b))
````

### Problem: Print a Circle

Implement a function named `circle_print(r)` that accepts
a single argument giving a radius,
and prints out a filled circle using asterisks.
Assume that each asterisk is square,
and of unit height and width.

````python
# Put your code here.
def circle_print(r):
  r1 = r**2
  final = ''
  for i in range(2*r):
    y = (i - r)**2
    for j in range(2*r):
        x = (j - r)**2
        if x + y <= r1:
            final = final + '*  '
        else:
            final = final + '   '
    final = final + '\n'
  print(final)
r = int(input("Enter radius: "))
circle_print(r)
````

After simplification, the numerator and denominator should have no factors in common, other than 1.
You may use recursion or iteration to implement this.

````python
# Put your code here.
def simplify(f):
  c = 2
  while c < min(int(f[0]),int(f[1])) + 1:
    if int(f[0])%c ==0 and int(f[1])%c ==0:
      f1 = int(int(f[0]) / c)
      f2 = int(int(f[1]) / c)
      f = (f1,f2)
    else:
      c=c+1
  return (f1,f2)
num = int(input("Enter numerator: "))
den = int(input("Enter denominator: "))
f = (num,den)
print(simplify(f))
````
