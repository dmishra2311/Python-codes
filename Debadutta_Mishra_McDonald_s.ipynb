```python
#Import required packages
import numpy as np
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.2f' % x)
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#Read the file
df = pd.read_csv("Mcdonald.csv")
df.head()
```

```python
#Q1
from matplotlib.pyplot  import figure
figure(figsize=(10, 10), dpi=80)
sns.countplot(df["Category"], order = df["Category"].value_counts().index)
plt.xticks(rotation='45')
plt.show()
```

Q1
> We can infer from the above plot, that 'Coffee & Tea' have the most varieties. Whereas 'Salads' have the least of varieties.

> We can also assume from the above plot that the varieties of  'beverage' and 'Chicken & Fish' are almost the same.

> We can also assume that there is a good demand for the category 'Coffee & Tea', hence so many varieties and hence could be one of the prime contributors to Revenue generation.

```python
# Assigning measure columns to a variable
columns1 = df.columns
columns1 = columns1.drop('Category')
columns1 = columns1.drop('Item')
columns1
```

```python
# Describing the data
df.describe().T
```

```python
#Changing 'Serving Size' to int datatype
#Q2
df['Serving Size'] = df['Serving Size'].str.replace(r'\D', '').astype(int)
for i in columns1:
    sns.boxplot(df[i])
    plt.show()
```

```python
#Q3
dfcorr = df.corr().abs()
dfs = dfcorr.unstack()
dfsort = dfs.sort_values(kind="quicksort")
dfsort.sort_values(ascending = False).head(50)
```
